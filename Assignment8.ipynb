```python
import string
import math

# Task 1
s = "Hello"
print(s.lower())

# Task 2
s = "HeLLo WoRLd"
print(s.swapcase())

# Task 3
s = "HelloWorld"
print(''.join([ch for ch in s if not ch.isupper()]))

# Task 4
s = "EngiNEEr"
upper_count = sum(1 for ch in s if ch.isupper())
lower_count = sum(1 for ch in s if ch.islower())
print(f"Uppercase: {upper_count}, Lowercase: {lower_count}")

# Task 5
s = "Data-Driven@2025!"
print(''.join([ch for ch in s if ch.isalpha()]))

# Task 6
a, b, c = 3, 4, 5
s_heron = (a + b + c) / 2
area = math.sqrt(s_heron * (s_heron - a) * (s_heron - b) * (s_heron - c))
print(area)

# Task 7
names = ["Alice", "Bob", "Charlie"]
max_len = max(len(name) for name in names)
for name in names:
    print(name.ljust(max_len), "|")

# Task 8
s = "   Hello, World!    "
s = s.strip()
s = ''.join(ch for ch in s if ch.isalnum())
s = s.replace(' ', '')
print(s)
```
