```python
import random

# Generate a random number between 0 and 7
correct_number = random.randint(0, 7)
guesses = 0

print("Guess a number between 0 and 7:")

while True:
    try:
        guess = int(input("Enter your guess: "))
    except ValueError:
        print("Please enter a valid number!")
        continue

    guesses += 1  # Count each guess

    if guess == correct_number:
        print(f"Correct! The number was {correct_number}.")
        print(f"You guessed it in {guesses} tries.")
        break
    else:
        print("Incorrect, try again!")
```
