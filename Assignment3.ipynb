```python
def get_name():
    """Ask for and return the user's name."""
    name = input("Enter your name: ")
    return name

def ask_name(name):
    """Ask what the user's name is and print it."""
    print("What is your name?")
    print(f"My name is {name}.")

def get_problem():
    """Ask the user what their problem is."""
    problem = input("What is your problem? ")
    return problem

def check_even_odd(name):
    """Ask for a number, then print whether it's even or odd, showing the user's name first."""
    number = int(input("Enter a number to check if it's even or odd: "))
    print(f"{name}, here is your answer:")
    if number % 2 == 0:
        print(f"{number} is an even number.")
    else:
        print(f"{number} is an odd number.")

# Main program flow
def main():
    name = get_name()
    ask_name(name)
    get_problem()
    check_even_odd(name)

# Run the program
main()
```
