```python
def multiplication_table(number, end):
    """Generate and print the multiplication table for the given number up to 'end'."""
    print(f"\nMultiplication Table for {number} up to {number} x {end}:\n")
    for i in range(1, end + 1):
        print(f"{number} x {i} = {number * i}")

def main():
    """Main function to take user input and call the table generator."""
    try:
        num = int(input("Enter a number to generate its multiplication table: "))
        end = int(input("Enter the number up to which you want the table: "))
        multiplication_table(num, end)
    except ValueError:
        print("Please enter valid integers.")
```
