```python

import random

def get_random_number():
    """Generate a random number between 0 and 7."""
    return random.randint(0, 7)

def get_guess():
    """Ask the user for a guess and return it as an integer."""
    while True:
        try:
            guess = int(input("Enter your guess (0 to 7): "))
            if 0 <= guess <= 7:
                return guess
            else:
                print("Please enter a number between 0 and 7.")
        except ValueError:
            print("Please enter a valid number.")

def check_guess(guess, correct_number):
    """Check if the guess is correct and return True if it is, False otherwise."""
    if guess == correct_number:
        print("Correct!")
        return True
    else:
        print("Incorrect, try again!")
        return False

def guessing_game():
    """Main game logic."""
    correct_number = get_random_number()
    attempts = 0

    print("Guess a number between 0 and 7:")

    while True:
        guess = get_guess()
        attempts += 1
        if check_guess(guess, correct_number):
            print(f"You guessed it in {attempts} tries.")
            break
```
