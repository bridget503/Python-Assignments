```python
import string
import math

# Task 1: Convert uppercase letters to lowercase
def to_lowercase(s):
    return s.lower()

# Task 2: Swap case for each letter
def swap_case(s):
    return s.swapcase()

# Task 3: Remove all uppercase letters
def remove_uppercase(s):
    return ''.join([ch for ch in s if not ch.isupper()])

# Task 4: Count uppercase and lowercase letters
def count_case(s):
    upper_count = sum(1 for ch in s if ch.isupper())
    lower_count = sum(1 for ch in s if ch.islower())
    return f"Uppercase: {upper_count}, Lowercase: {lower_count}"

# Task 5: Remove non-English letters
def remove_non_letters(s):
    return ''.join([ch for ch in s if ch.isalpha()])

# Task 6: Compute area of triangle using Heron's formula
def triangle_area(a, b, c):
    s = (a + b + c) / 2
    area = math.sqrt(s * (s - a) * (s - b) * (s - c))
    return area

# Task 7: Format a list of names in a neat table
def format_names(names):
    max_len = max(len(name) for name in names)
    for name in names:
        print(name.ljust(max_len), "|")

# Task 8: Clean a string
def clean_string(s):
    s = s.strip()  # Remove leading/trailing whitespace
    s = ''.join(ch for ch in s if ch.isalnum())  # Remove punctuation
    s = s.replace(' ', '')  # Remove spaces
    return s


# Example usage:
print(to_lowercase("Hello"))               # Task 1
print(swap_case("HeLLo WoRLd"))            # Task 2
print(remove_uppercase("HelloWorld"))      # Task 3
print(count_case("EngiNEEr"))              # Task 4
print(remove_non_letters("Data-Driven@2025!"))  # Task 5
print(triangle_area(3, 4, 5))              # Task 6
format_names(["Alice", "Bob", "Charlie"])  # Task 7
print(clean_string("   Hello, World!    "))# Task 8
```
